##### What is Accessibility
Accessibiliy is to make your website easy to interpret for all types of people or people who may require assistive technology 

* Examples
    * Screen Readers - Sight impairments
    * Screen magnifiers - Also sight impairments
    * Switch controls  - Motor controll impairments
    * Closed caption - Auditory processing and hearing impairments

Making websites available to all people is required. 

#### Web Content Accessibility Guidelines
    * Four principles or P.O.U.R
        * Perceivable
        * Operable
        * Understandable
        * Robust

#### WCAG Principles: Perceivable

* All non-text elements have a text alternative
    * examples: Images have a brief description

* Time Based Media 
    * Captions and transcripts for those who are hearing impairments
    * All pre-recorded video and audio must have captions and transcripts or closed captions

* Present content using semantic HTML
    * Using semantic html elements properly
    * Presenting HTML in a logical order

* Portrait and landscape orientation
    * Mobile designers and developers must make sure content can be understood and navigated in landscape and potrait mode.

* Sufficient color contrast
    * Text must have enough contrast with the backround

* Combine multiple visual cues, not just color

* Avoid autostarting audio or video

* Responsive body copy
    * Keep text at about 80 characters per line
    * maintain sufficient line height

* Legible at 200% zoom

#### WCAG Principles: Operable

* Keyboard Navigation
    * All interactive elements are accessable using the tab key
    * Avoid keyboard locking

* Obvious focus indicators
    * Selected obects must be visually obvious
    * 44 x 44 pixels for buttons

* Allow enough time 
    * When completing a task, users should be able to adjust or cancel a time limit

* Disable distracting content
    * such as discords ability to disable moving images

* Careful with flashing animations
    * videos and other things must not flash 3 times within a second

* Navigatable
    * users should be able to navigate find content and determine where they are
    * A mechanism should be available to skip blocks of content

* Descriptive page \<titile>
    * Make link purpose clear

* Semantic heading structure 
    * h1 element is reserved for the heading of the article

* Provide alternatives to complex interactions

* Generous target size

#### WCAG Principles: Understandable

* Developers should set a HTML language attribute

* Use plain language - clear and straightforward
    * Strong, simple words
    * Short paragraphs
    * Clearly organized information
    * A conversational tone

* Create predictable navigation

* No unexpected changes in context
    * The following should happen only when a user initiates a change by interacting with an interface element
        
        * Opening a new window
        * Loading a new page
        * rearranging page content
        * Moving focus to a new component

    * Make errors easy to spot and correct

        * Make messages easy to notice and understand
        * highlight errors visually
        * Keep instructions simple
