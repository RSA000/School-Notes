### Introduction To The Terminal (CLI)

* The terminal is also known as a CLI (command line interface)
* More powerful then a GUI (Graphic User Interface)
* Many Servers only allow command line access
* Focusing on UNIX based operating systems since they are the most popular

* Many terminals run a program called a shell
* For example, my computer runs zsh when I open my terminal
* tutorial is in bash but is very similar


##### Common Terminal Commands
* ls: 
    * lists all files in working diretory

* whoami: 
    *returns current user name

* cat: 
    * Joins contents of two files
    * example
         ```
         cat status.txt
         "A statue of a hunter standing over a dead bear. Creapy <-- Contents of status.txt
         ```
    * cat can take multiple arguments and will concatenate one file after the other in the order they where written

* CTRL + C:
    *  Keyboard Interrupt, quits current task

* clear or CTRL + L:
    * Clears the screen

* cd:
    * stands for change directory, allows user to change current working directory

* echo:
    * echo takes arguments and converts them into output
    * example
    ``` 
    echo "hello  There"
    output: hello  There
    quotations can be used group multiple words into a single argument
    ```


###### Options
Options are special 'addons' to command that change how they function.
for example ls -a will display all files and hidden files (example .gitignore) in a current directory. Another example is ls -t which will list the files based on the time they where created

* Most commands can have multiple options (example: ls -a -t)
* commands options can be longer with a -- in front of them (example: ls --all = ls -a)

##### Shortcuts
* history:
    * will show previous commands entered
    * can also be accessed one by one with the up and down arrows
* tab can also be used as an autocomplete in some instances
