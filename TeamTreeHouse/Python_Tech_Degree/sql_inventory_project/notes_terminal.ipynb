### Introduction To The Terminal (CLI)

* The terminal is also known as a CLI (command line interface)
* More powerful then a GUI (Graphic User Interface)
* Many Servers only allow command line access
* Focusing on UNIX based operating systems since they are the most popular

* Many terminals run a program called a shell
* For example, my computer runs zsh when I open my terminal
* tutorial is in bash but is very similar


##### Common Terminal Commands
* ls: 
    * lists all files in working diretory
    
    * You can add arguments such as (ls directory) to show the contents of a directory you are not in

* whoami: 
    * returns current user name

* cat: 
    * Joins contents of two files
    
    * example
         ```
         cat status.txt
         "A statue of a hunter standing over a dead bear. Creapy <-- Contents of status.txt
         ```
    
    * cat can take multiple arguments and will concatenate one file after the other in the order they where written

* CTRL + C:
    *  Keyboard Interrupt, quits current task

* clear or CTRL + L:
    * Clears the screen

* cd:
    * stands for change directory, allows user to change current working directory

* echo:
    * echo takes arguments and converts them into output
    * example
    ``` 
    echo "hello  There"
    output: hello  There
    quotations can be used group multiple words into a single argument
    ```
* pwd:
    * Stands for print working directory. Shows current directory user is in and path to that directory.
    * example output: /home/user/etc...

* less
    * A pager program
    * convienent way to view just some of the text in a file
    * like more but can also scroll with arrow keys
    * doesn't automatically quit when reaching the bottom

* more
    * like less, user can press space to display more
    * automatically quits when bottom of the file is reached

* cp
    * Allows user to copy files and directories
    * syntax (file_to_copy.txt file_to_copy_to.txt)
    * You can also copy into other directories (file_to_copy /somedirectory)
    * You can also copy directories into directories (cp -r some_directory/ another_directory)
        * the -r is needed and recursively copies

* mv 
    * allows user to move files or directories to other files or directories (some_file some_directory)
    * also can be used for renaming files (file_to_change file_to_change to)
    * You can also apply the wildcard to move all files with .txt for example

* rm 
    * used to remove files
    * is very permenent
    * syntax (rm some_file_or_directory)

* mkdir 
    * is used to make new directories 

* find
    * used to recursively search through a directory and all its subdirectories for files
    * You can also limit the results to files whose names matches a particular patten with -name option

* grep 
    * command is used to print lines in a file that match a particular pattern 

##### Options
Options are special 'addons' to command that change how they function.
for example ls -a will display all files and hidden files (example .gitignore) in a current directory. Another example is ls -t which will list the files based on the time they where created

* Most commands can have multiple options (example: ls -a -t)
* commands options can be longer with a -- in front of them (example: ls --all = ls -a)

##### Shortcuts
* history:
    * will show previous commands entered.
    * can also be accessed one by one with the up and down arrows.
* tab can also be used as an autocomplete in some instances.


### Directories and Files 
* Pretty much everything in a computer is either a file or a directory. Directories contain files and files contain data.

* Directories use a parent-child type of classification. If a directory is in another directory, the directory is a child of the first directory and the directory with the directory inside of it is the parent directory.

* When using pwd, directories are separated with /.

* . refers to whatever directory you are in. .. refers to the directory that your current directory is in.


##### Relative Paths
* A relative path is a file path that is relative to your starting location in a file system

* A relative path that works in one directory may not work in another

##### Absolute Paths
* Paths the start from the root (base directory represented with a /
* absolute Paths always start from the / directory and can work from any directory


##### Directory Shortcuts
* '.' indicates current directory
    * One use of . is with executable programs
    * for an executable to be run it needs to have its path specified (directory/exe.sh)
    * You can use . if the executable is within your current directory

* '..' Represents the parent directory
    * can be used to navigate one directory up (cd ..)
    * or to navigate a path (cd ../some_directory)

* ~ (tilde) represents users home directory 
    * can be used to navigate to users home directory from anywhere (cd ~) instead of typing out an absolute path or relative path

##### Wildcard Expansion

* '*' The asterisk is used for wildcard expansion, where a pattern is used to complete one or more file or directory names
    
    * example: To select all files ending with ".txt" you could run (cat some_directory/*.txt)
    
    * syntax: file_name_start*file_name_end

    * '*' can be used on its own to select everything within a directory

    * (ls */) will list all the directories files and directories within your directory


##### File and Directory Naming Conventions
* file names are case sensitive on a pc
* macs use case insentive
* regardless, all directories should be lower case
* stick to ascii characters
* numbers are fine
* punctuation should be avoided
* spaces should be avoided
