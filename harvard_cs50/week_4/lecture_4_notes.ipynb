### Dynamic Memory Allocation. Malloc and free

Malloc - Memory Allocate
    * How much memory you want to utilize

Free - Hands memory back to operating system. 
    * Frees data.

example code
```c
#include <cs50h>
#include <ctype.h>
#include <stdio.h>
#include <string.h>
#include <stdlib.h>

int main(void)
{
    // creating an array of characters
    char* s = get_string("S: ");

    // allocating memory of size of s
    // +1 to account for \n
    char* t = malloc(strlen(s) + 1);
    // handling a case where there is not enough memory
    if (t == NULL)
    {
        return 1;
    }

    // for loop to copy string
    for (int i = 0, n = strlen(s) + 1; i < n; i++)
    {
        t[i] = s[i]; 
    }

    // handling if t is of 0 length (empty)
    if (strlen(t) > 0)
    {
        t[0] = toupper(t[0])
    }
    

    printf("s: %s\n", s);
    printf("t: %s\n", t);

    // freeing to and making memory available
    free(t);
    return 0
}

// output
// hi
// hi
// Hi
```

* sometimes malloc can fail if there is not enough memory to allocate

* NULL is used to check if a pointer is valid or not

### Binky Pointer Fun

bink.c
```c
int main(void)
{
    // declaring variables
    int *x;
    int *y;

    // allocating memory to x
    x = malloc(sizeof(int));

    *x = 42;
    // This could cause problems because we have not allocated memory to y
    *y = 13;

    // y is now pointing to x
    y = x;

    // y is now equal to 13 and so is x? 
    *y = 13;
}
```

swap.c 
```c
#include <stdio.h>

void swap(int a, int b);

int main(void)
{
    int x = 1;
    int y = 2;

    printf("x is %i, y is %i\n", x, y);
    swap(x, y)
    printf("x is %i, y is %i\n", x, y)
}

void swap(int a, int b)
{
    int tmp = a;
    a = b
    b = tmp
}
```

There's is a bug here. a and b are local to the function so x and y are not actaully used. We use copys of them. To fix this, we can pass in the pointers of x and y

```c

#include <stdio.h> 

void swap(int *a, int *b);

int main(void)
{
    int x = 1;
    int y = 2;

    printf("x is %i, y is %i\n", x, y);
    swap(&x, &y)
    printf("x is %i, y is %i\n", x, y)
}

void swap(int *a, int *b)
{
    int tmp = *a;
    *a = *b
    *b = tmp
}

scanf.c getting an integer with scanf
```c
#include <stdio.h>

int main(void)
{
    int x;
    printf("x: ");
    scanf("%i", &x);
    printf("x: %i\n", x);
}
```
