#### Pointers
A pointer is a variable that stores the address of a value.

* A single **&** is the "address of" operator 
* If you want to store an adress in a variable you have to use * 

example to see pointer:
```c
int main(void)
{
    #include <stdio.h>

    int n = 50;
    int *p = &n;
    printf("%p\n", p);

    //output in lecture: 0x7ffcb4578e5c
}
```

You could also do:

```c
#include <stdio.h>

int main(void)
{
    int n = 50;
    printf("%p\n", &n);
}
```
and achieve the same result

* The * can also be used to print out the value of an address:

```c
#include <stdio.h>

int main(void)
{
    int *p = &n;
    int n = 50;
    
    // should print n's 
    printf("%p\n", p);
    // should print 50
    printf("%i\n", n)
}
```

```c
#include <stdio.h>

int main(void)
{
    int *p = &n;
    int n = 50;
    
    printf("%p\n", p);
    // go to address
    printf("%i\n", *p)
    // should return 50
}
```
