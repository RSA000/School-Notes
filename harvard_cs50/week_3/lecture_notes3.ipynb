### Algorithms

* An array is contiguous memory (as in it's contained byte after byte after byte)



##### Searching 

* Searching is one of the most implemented uses for a computer today. 

* Determining a algorithms running time is used to find how efficient a algorithm is. 

* A linear search is searching element by element

### Big O Notation

* Big O notation is used to determine the time to solve over the size of the problem.

##### Common Big O Notations
* O(n^2)
* O(n log n)
* O(n) - linear search
* O(log n ) - binary search 
* O(1)
* The higher the mathematical expression the slower the algorithm (more steps)

* Omega is used to describe a lower bound on the time of an algorithm.
* Big O describes an upper bound
* Big theta is used when an upper bound and lower bound are the same.


### Implementing Linear Search

Linear search with numbers
```c
#include <cs50.h>
#include <stdio.h>

int main(void)
{
    int numbers[] = {4, 6, 8, 2, 7, 5, 0};

    for (int i = 0; i < 7; i++)
    {
        if (numbers[i] == 0)
        {
            printf("Found\n");
            return 0;
        }
    }
    printf("Not Found\n");
    return 1;
}
```

Linear search with letters
```c
#include <cs50.h>
#include <stdio.h>
#include <string.h>

int main(void)
{
    string names[] = {"Bill", "Charlie", "Fred", "George", "Ginny", "Percy", "Ron"}

    for (int i = 0; i < 7; i++)
    {
        if (strcmp(names[i], "Ron") == 0)
        {
            printf("Found\n");
            return 0;
        }
    }
    printf("Not Found\n");
    return 1;
}
```
