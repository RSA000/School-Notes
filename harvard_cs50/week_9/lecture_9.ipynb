### Flask

Flask is a Python library that allows us to serve websites and deal with user data.

routes are something like http://webexample.com/route

input = http://webexample.com/route?key=value - using GET

Flask is a framework - a bunch of libraries for people to use


using Flask

* app.py - Actual code 
* requirements.txt - required libraries/python version
* static/ - files that aren't going to change
* templates/ - Any html


##### app.py


example

```python

from flask import Flask, render_template, request

# __name__ refers to the name of the current file
app = Flask(__name__)


@app.route("/")
def index():
    name = request.args.get("name")
    return render_template("index.html", name=name)

# A decorator is a special type of function that modifies another function
```

```html
<!DOCTYPE html>

<html lang="en">
    <head>
        <meta name="viewport" content="initial-scale=1, width=device-width">
    </head>
    <body>
        hello, {{ name }}
    </body>
</html>
```

flask run - creates url


```python

from flask import Flask, render_template, request

# __name__ refers to the name of the current file
app = Flask(__name__)


@app.route("/")
def index():
    return render_template("index.html")

@app.route("/greet")
def greet():
    name = request.args.get("name")
    return render_template("greet.html", name=name)

```

greet.html

```html
<!DOCTYPE html>

<html lang="en">
    <head>
        <meta name='viewport' content='initial-scale=1, width=device-width'>
        <title>greet</title>
    </head>
    <body>
        hello, {{ name }}
    </body>
</html>

```

Layout.html

```html

<!DOCTYPE html>

<html lang="en">
    <head>
        <meta name='viewport' content="initial-scale=1, width=device-width">
        <title>hello</title>
    </head>
    <body>
    {% block body %}{% endblock %}

    </body>
```

index.html

```html

{% extends "layout.html" %}

{% body block %}

<form action="/greet" method="get">
    <input autocomplete="off" autofocus name="name" placeholder="Name" required type="text">
</form>

{% endblock %}

```

### GET vs POST

Get puts key:value pairs into the url, post does not

Why not use post everywhere? 
    * Your history is less useful, you'd have to refill out the form

UX - user experience

request.args - arguements in url

requests.forms - arguements in form
