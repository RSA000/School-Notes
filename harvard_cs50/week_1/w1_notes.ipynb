#### Learning a little bit of C

* C is a general purpose langauge created in the 1970s by Dennis Ritchie.
* C was developed at Bell Labs in the 1970s
* C is a compiler based program 
* ; is used as a statement seperator and {} are used for grouping blocks of statements
* Keywords: if/else, for, do/while, while and switch
* arithmetic operators: + += ++/--, &, ||, %


##### An example of a basic C program
```c
#include <stdio.h>
int main(void)
{
    printf("hello, world\n")
}

```

To actually run this code we need to compile it. A compiler is like a translator, it takes our code and translates it into something our computers can understand and run. For example, we can compile our code with gcc which will create an executable that will run our code. The executable is machine language of our C langauge 

* source code is that we can read or write.

##### Loops
basic example of a C while loop:
```c
while (true)
{
    printf("meow\n")
}
```
The above code would print 'meow' infinetly.

example of a basic for loop:
```c
for (int i = 0; i < 3; i++)
{
    printf("meow\n")
}
```
The above will print out "meow" three times
* i++ is and syntactical way of writing i = i + 1

You could also write it like this:
```c
int i = 0;
while (i < 3)
{
    printf("meow\n");
    i = i + 1
}
```

#### Functions
You can also create blocks of code and call them later:
```c
#include <stdio.h>

// This is used to not throw an error since our functions are usually at the bottom
// our meow takes a parameter that is an integer, n
void meow(int n);

// Here we have a loop that will call the function in our main
int main(void)
{
    meow(3);
}

void meow(int n)
{
    // We create i which is an integer equal to zero and while
    // i is less than n (our parameter) the loop will call our meow function and increment i by 1
    for (int i = 0; i < n; i++)
    {
        printf("meow\n");
    }
}
```

#### Floating point imprecision and Integer overflow
Computers store numbers in bits, so if a number is larger then what the computer can store, problems can occur. In math, there is an infinite number of floating point numbers but with computers you can only store numbers in a finite amount of space (like 32 bits).<br>

So something like 1/10 would cause some problems due to the way the binary system works.

Integer overflow occurs when you use a numbers that higher then the memory that can contain a number. For 32 bits that would be 2,147,483,647
or (01111111111111111111111111111111). The first bit is known as a sign bit and indicates whether a number is positive or negative.
