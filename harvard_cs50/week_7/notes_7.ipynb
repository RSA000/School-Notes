## SQL


We Usally rely on databases to store information on a website.

* A database is a hierachically organized set of tables comprised of rows and columns

* SQL Stands for Structure Query Language and is used to query a database

* MySQL is one open-source platform which you can esablish the type of relational database that SQL is most adept at working with
    * SQLite is another, which we've actually used in CS50 since 2016

* Many installations of SQL come with a GUI tool call phpMyAdmin which can be used to execute database queries in a more user-friendly way.
    * Used to build databases in the first place usally



* Unlike in C, CHAR refers to a fixed-length string
    * In most relational databses, you actually specify the fixed-length as part of the type definition. ex: CHAR(10) - Must be exactly 10 characters

* VARCHAR refers to a variable-length string.
    * VARCHARs also require you to specify the maximum possible length of a string that could be stored in that column. e.g. VARCHAR(99) - 0 - 99 characters

While there are many data types, they all have a 'type affinity' which can relate to the one of the following:
    * NULL
    * INTEGER
    * REAL
    * TEXT
    * BLOB

One column should be a a primary key - which should be able to be uniquely and quickly identified

INSERT
    * Add information to a table
    
```sql
    INSERT INTO
    <table>
    (<columns>)
    VALUES
    (<values>)


INSERT INTO
users
(username, password, fullname)
VALUES
('newman', 'USMAIL', 'Newman')
```


* When defining the column that will be your table's primary key, its usually best practice for it to be an integer

* You can configure the primary key to autoincrement so it will automatically pre-populate that column for you automatically when rows are added to the table.


SELECT
    * Extract information from a table

```SQL 
SELECT
<columns>
FROM
<table>
WHERE
<condition/predicate>
ORDER BY
<column>
```

SELECT * - means every column

* Databases empower us to organize information into tables efficiently.
    * We don't always need to store every possible relevant piece of information in the same table, but can use relationships across the tables to let us pull information from where we need it


SELECT (JOIN) 
    * Extract information from multiple tables

```sql
SELECT
<columns>
FROM
<table1>
JOIN
<table2>
ON
<predicate> 
```

UPDATE
    * Modify information in a table

```sql
UPDATE
<table>
SET
<column> = <value>
WHERE
<predicate>
```


DELETE
    * removes information from a table

```sql
DELETE FROM
users
WHERE
username = 'newman'
```
