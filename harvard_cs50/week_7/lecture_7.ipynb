#### Regular Expressions

Regular Expressions allow you to express patterns in a standardized way

examples:
    * .\* - any character multiple times
    * @ matches @ characters
    * \. - escape so that we can match .
    * .+ - any characters one or more times
    * - ? optional
    * ^ - start of input 
    * $ - end of input

```python

import csv
import re

counter = 0

with open("favourites.csv) as file:
    reader = csv.DictReader(file)
    for row in reader:
        title = row["title"].strip().upper()
        if re.search("^OFFICE|THE.OFFICE)$"), title):
            counter += 1
```
