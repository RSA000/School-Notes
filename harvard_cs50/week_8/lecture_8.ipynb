#### The Internet!

* A bunch of data we can all reach
* Servers interconnected via wires, wireless
* Interconnected Networks :D


Picture a bunch of connected dots on a map, the connections represent networks that are connected. A router routes data from one point to another.

TCP/IP - Refers to two protocals to intercommunicate
    * Protocal is like a set of rules

Steps: 
    * An 'envelope' holding information with a unique address on it
    * Also include a source address - IP address
    * IP Address #.#.#.# - 0 - 255 32 bits or four bytes - 4 billion IP addresses available
        * Nowadays, there's different types of IP since we are at the point of exceeding 4 billion addresses.

* Email servers, chat servers, web servers

* TCP - Allows us to specify the type of service
    :port
    common TCP: 80 - http - hypter text transfer protocal, 443 - HTTPS


##### Fragmentation
    * Fragments file into smaller pieces, fragments need order.
    * Able to determine if recieved or not entirely recieved with order



* IP handles the addressing
* TCP handles the standardization of what services use


##### DNS - Domain Name System
    * Pretty much every network on the internet has a DNS server
    * Contains a 'spreadsheet' containing Domain Name, and IP Address


##### HTTP/S - Hypertext Transfer Protocal / Secure
    A final protocal that determines how web browser and web servers intercommunicate

#### HTML - Hyper Text Markup Langauge

* Not a programming - doesn't express logic but more the structure of a page

Example of html code

```html

<!DOCTYPE html> - Document type declaration

<html lang="en"> - html tag, specifying it is in english which is an example of an attribute
                    known as an open tag
    <head> - head of the page
        <title>hello, title</title> - typically embedded in tab
    </head>
    <body> - body of page - 99% of webpage usually
        hello, body - body of page
    </body>
</html> - Closing tag

```

* TCP port 80 or 443 - http or https


* HTML is characterized by two features, tags, and attributes.

* head and body are siblings

* title is a child of head. Titles child is text


* <p> - paragraph
* <h1, h2, h..> - headings
* <ul> - unordered list
* <li> - item in list
* <ol> - Ordered list
* <table> - Create table
* <thead> - table head
* <tbody> - table body
* <th> - table heading
* <tr> - table row
* <td> - table data
* <img scr=""> - Image - src is source of image. alt describes image for visually impaired
* <video> <source src=:: type::> </video> - used for videos
* <iframe scr=""> </iframe> - used for embedding videos
* <a href=""> - a stands for ancor - used for linking
