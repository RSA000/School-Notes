### Internet Primer

We are now transitioning to the world of the internet!

Things needed to connect to the internet:

* IP Address - Internet Protocal

* An IP address uniquely identifies a machine

* The addressing schema used by computers is known as IP addressing

* At first was a unique 32 bit address
    * 4 Clusters of 8 bits

* IP address looks like x.x.y.z
    * Each of w, x, y, and z can be nonnegative value in the range [0, 255]
    * 123.45.210.1

* There are workarounds for IP addresses since we now have over 7 billion people on the planet and generally people can own multiple devices.


* In recent years, we've been slowly phasing out this old sheme (IPv4) and replacing it with a newer sheme (IPv6) that assigns computers 128-bit addresses, instead of 32-bit addresses
    * s:t:u:v:w:x:y:z
    * 8 clusters of 16 bytes - 128 bits
    * represented in hexidecimal


##### DHCP

* Somewhere between your computer and the Internet at large exists a Dynamic Host Configuration Protocol (DHCP) server, whose role is to assign IP addresses to devices.

* Prior to the widesperad promulgation of DHCP, the tast of assigning IP addresses fell to a system administrator, who would need to manually assign such addresses

##### DNS

* Usually, you've never actually tried to visit a website by typing it's IP address into your browser

* The Domain Name System (DNS) exists to help us translate IP addresses to more memorable names that are more human readable

* DNS is like a big phone book with connected names and numbers.

* Much like there is no yellow pages of the world, there is really no DNS record of the entire internet

* Rather, large DNS server systems (like Google) are more like aggregators, collecting small sets of DNS information and pooling them together, updating frequently.

* IN that way, large DNS servers are like libraries that stock many diddferent sets of local yellow page books.

* DNS record sets are thus fairly decentralized


##### Access Points

* One of the ways we've dealt with the IPV4 addressingn problem is to start assigning multiple people to the same IP address

* The IP address is assigned to a router, whose job it is to act as a traffic cop that allows data requests from all of the devices on your local network (home, business, ect..) to be processed through a single IP address.

* Modern home networks consists of access points that combine a router, a modem, a switch, and other technologies together into a single device

* Modern business networks or large-scale wide area networks (WANs) still frequently have these as separate devices to allow the size of their network to scale more easily

* The internet is really just a system of protocols

### IP - Internet Protocol

* The internet is an interconnected network comprised of smaller networks woven together and agreeing to communicate with one another

* These networks know how to communicate with one another with IP

* Networks are not connected to every other network
    * With only six networks, things would already get out of hand connecting them this way. 
    * We still need each network to be able to talk to each other network though, or we end up with pieces of the network that are unable to speak to other parts of the network

* This is where routers come back into play

* What if, instead of being connect to every other network, each network was connected to a limited number of routers (each of which was connected to other nearby routers), and each router had instructions built into on how to move information towards its destination?

* This information might be stored in a routing table, inside of the router

* Using routers, the networks are not directly connected to each other at all, and rely on routers to distribute communications

* On a small scale, the configuration may be less efficient than just having direct connections

* On a large scale, this configuration can dramatically reduce the costs of network infrastructure

* Instead of sending large amounts of data all at once, information is sent in packets

* IP is also known as a connectionless protocol. There is not necessarily a defined path from the sender to the receiver, and vice versa
    * This means that in response to traffic that might be "clogging" up one particular path through the internet, some packets can be "re-routed" around the traffic jam to follow the most optimal path, based on the current state of the network



### TCP - Transmission Control Protocol

IP - moves inforomation from sending to recieving machine

* If the Internet Protocol (IP) is thought of as the protocol for getting information from a sending machine to a receiving machine, then Transmission Control Protocol (TCP) can be thought of as directing the transmitted packet to the correct program on the receiving machine.

* As you might imagine, it is important to be able to identify both where the receiver is and what the packet is for, so TCP and IP are almost inseperable pair: TCP/IP


* Each program/utility/service on a machine is assigned a port number. Coupled with an IP address, we can now uniquely indentify a specif program on a specific machine

* The other thing that TCP is crcial for is guaranteeing delivery of packets, which IP alone does not do.

* TCP does this by including information about how many packets the receiver should expect to get, and in what order, and transmitting that information alongside the data.


Some ports are so commonly used that they have been standardized across all computers
    * FTP (file transfer) uses port 21
    * SMTP (email) uses port 25
    * DNS - uses port 53
    * HTTP (web browsing) uses port 80
    * HTTPS (secure browsing) uses port 443


##### Steps of the TCP/IP process

* When a program goes to send data, TCP breaks it into smaller chunks and communicates those packets to the computer's network software, adding a TCP layer onto the packet.

* IP routes the individual packets from sender to receiver; this info is part of the IP layer surrounding the packet.

* When the destiniation computer gets the packet, TCP looks at the header to see which program it belongs to; and since the routes of packets may differ, TCP also must present those packets to the destination program in the proper order


* If at any point along the way a router delivering information using the internet protocol dropped a packet, TCP would use additional information inside the headers to request that the sender pass along the extra packet so it could complete assembly

* After the packets have arrived, TCP ensures they are organized in the correct order and can be reassembled into the inteded unit of data and delivered to the correct service.



### HTTP - Hyper Text Transfer Protocol

* In addition to protocols that dictate how information is communicated from machine to machine and application to application (IP and TCP) It is requently the case that the application itself has a system of rules for how to interpret the data that was sent

* HTTP is one such example of an appliaction layer protocol, which specifically dictates the format by which clients request web pages from a server, and the format via which servers return information to clients.

* Other appliaction layer protocols include:
    * File transer Protocol (FTP)
    * Simple Mail Transfer Protocol (SMTP)
    * Data Distribution Service (DDS)
    * Remote Desktop Protocol (RDP)
    * Extensible Message and Presence Protocol (XMPP)

example: 

We might send:

GET / HTTP/1.1
Host: cats.com


and recieve:

HTTP/1.1 200 OK
Content-type: text/html

* Lines of the form:
    * HTTP request line:  method request-target http-version


* Based on whether the resource exists and whether the server is empowered to deliver that resource pursuant to the client's request, a number of status codes can result.

Types of codes

* 200 - Success 
* 301 - Moved Permenently, page is at a new location
* 302 - Found, Page is moved temporarily
* 401 - Unauthorized - page requires some kind of authorization such as logging in
* 403 - Forbidden - Server will not allow this request
* 404 - Not found -- file doesn't exist on server
* 500 - Internal Server Error - Server failing to deliver request somehow

### CSS - Cascading Style Sheets

* CSS Is another language we use when constructing websites
    * If HTML is used to organize the content that we aim to display on our pages, then CSS is the tool we use to customize our website's look and feel

* Like HTML, CSS is not a programming language; it lacks logic. Rather, it is styling language and its syntax describes how certain attriuvutes of HTML elements should be modified

simple CSS stylesheet

```CSS
body
{
    background-color: blue;
}
```

* A style sheet is constructed by identifying a selector and then open curly braces to indicate thebeginning of the style sheet for that selector

* In between the curly brace you place a list of key-value pairs of style properties and values for those properties, each declaration ending with a semicolon

* Then closing curly brace terminates the style sheet


Common CSS properties 

* border: style color width
    * Applies a border of the specified color, width, and style (e.g. dotted dashed etc..)

* background-color: [keyword | #<6-digit hex>]
    * Sets the background color. Some colors are pre-defined in CSS.

* color: [keyword | $<6-digit hex>]
    * Sets the foreground color (usually text)

* font-size [absolute size | relative size]
    * Can use keywords [xx-small, medium...), fixed pints (10pt, 12pt..)]
    percentage (80%, 120%), or base off the most recent font size (smaller, larger)

* font-family: [font name | generic name]
    * Certain "web safe" fonts are pre-defined in CSS

* text-align: [left | right | center | justify]
    * for displaying text


* Your selectors don't have to apply only to HTML tag categories. There also exists ID selectors and class selectors

* A tag selector will apply to all elements with a given HTML tag

```CSS 

/* tag attribute */
```



### DOM - Document Object Model

* As we've seen, JavaScript objects are incredibly flexible, and can contain various fields, even when those fields are other objects

* The document object is one way of employing this paradigm, whereby that object organizes the entire contents of a web page.

* By organizing an entire page into a JavaScript object, we can manipulate the page's elements programmatically

example 

```html

<html>
    <head>
        <title> Hello, world</title>
    </head>
    <body>
        <h2>Here's my page</h2>
        <p>World, hello</p>
        <a href="test.html">Link</a>
    </body>
</html>

```

* The document object itself, as well as all of the objects contained within it, have a number of properties and a number of methods that can be used to drill down to a very specific piece of your website.

* By resetting those properties or calling certain methods, the contents of our web pages can change without us needing to refresh the page.

console.dir - tries to layout page like directories

* DOM properties
    * innerHTML - Holds the HTML inside of HTML tags
    * nodeName - The name of an HTML element or element's attribute
    * id - The "id" attribute of an HTML element
    * parentNode - A reference to the node one level up in the DOM 
    * child Nodes - An array of references to the nodes one level down in the DOM
    * attributes - An array of attributes of an HTML element
    * style - An objects encapsulating the CSS/HTML styling of an element


* Dom Methods 
    * getElementById(id) - Gets the element with a given ID below this point in the DOM
    * getElementsByTagName(tag) - Gets all elements with the given tag below this point in the DOM
    * appendChild(node) - Add the given node to the DOM below this point
    * removeChild(node) - Remove the specified child node from the DOM

* If we start from document, we can get any piece of our web page that we choose, through careful use of DOM properties and methods
