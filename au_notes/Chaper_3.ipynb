* Efficiency is the term used to describe an algorithm's careful use of resources.

* Algorithms can be measured in how much time and space they take up.




##### Measuring effiency

* The study of the effiency of algorithms is called the analysis of algorithms.

Sequential search is an algorithm that loops through a list looking for a particular value.

```python 

# Example of a sequential search algorithm

# List to be searching through 
list = ['dog', 'cat', 'bat']

# search function takes a word to search and a list as parameters
def search(word, list):
    for word in list:
        # if word matches, print 'found' and break out of loop
        if word == word:
            print('found')
            break
    # If the list is exhausted, print 'not found'
    print('not found')
```

This algorithms effiency can be measured on how many comparisons it may need to do.

For example, with the above algorithm, the function could loop through three times if the word is not in the list or if the word is at the end. The function could also find the word at the beginning resulting in only 1 comparison. The average would be 2 comparisons. 

The work required is proportionate with the size of the list. As the size of the list increases, the amount of work also increases at a constant rate in a worst case scenario. 

There is also space effiency to note. There is the memory required for the list and then memory required for the comparison and the word to be found. Overall, this doesn't require a huge amount of memory.

* Sequential search is O(n) where n is the size of the list. In the worst case scenario, it would take n comparisons (the entire length of the list) to complete. 





#### Order Of Magnitude : Order n

* O(n) Work increases with size of n

* O(n)2 Work increases exponentially with increase of n

* O(log n) Work increases to the log of n


If you were given two algorithms to choose from that performed the same task, it is usually the best choice to choose the algorithm with the lowest order of magnitude.
