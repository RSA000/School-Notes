### Computer Systems Organizations

* The branch of computer science that studies computers in terms of their major functional units is computer organization. 

* Hierachy of abstractions.

##### The Components of a Computer System.

The Von Neumann architecture is based on the following three characteristics.

* Four major subsystems called memory, input/output, the arithmetic logic unit (ALU), and the control unit. These four units are often bundled together as a Central Processing Unit or CPU

* The stored program concept, in which the instructions to be executed by the computer are represented as binary values and stored in memory.

* The sequential execution of instructions, which one instrution at a time is fetched from memory and passed to the control unit, where it is decoded and executed.

#### Memory and Cache

* Memory is the functional unit of a computer that stores and retrieves the instructions and the data being executed. All information stored in memory is represented in binary. 

* Computer memory uses an access technique called random access, and the memory unit is frequently referred to as Random access memory (RAM)

RAM has the following three characteristics

* Memory is divided into fixed-size units called cells, and each cell is associated with a unique identifier called an address. These addresses are the unsigned integers 0, 1, 2,.. 

* All accesses to memory are to a specified address, and we must always fetch or store a complete cell- that is, all the bits in the cell. The cell is the minimum unit of access.

* The time it takes to fetch or store the contents of a cell is the same for all the cells in memory.

* In a cell, the number of bits per cell is called the cell size or the memory width and is usually denoted as W.

* The standard cell size is 8 bits or 1 byte.

* Often computers use 2 or 4 bytes to store a single data value. This gives the range need, but requiress more trips to memory rather that a single fetch.

* If there are N bits available to represent the address of a cell, then the smallest address is 0 and the largest address is a string of N 1s:
Which is equal to the value 2^n - 1

* The value 2^n is called the maximum memory size or the address space of the computer. 

* The basic memory operations are fetching and storing. 

    * Nondestructive fetch - value remains in orginal address

    * Destructive store - overwriting something.

* Memory access time - The time it takes to fetch or store data in ram

* A cell is the smallest accessible unit of memory

* ROM - Read Only Memory

* Memory registers are needed to perform fetch and store operations.
    * Memory Address Register (MAR) - holds the address of the cell to be accessed. Must be N bits wide, where 2^n is the address space of the computer

    * Memory Data Register (MDR) - Contains the dfata value being fetched or stored.


#### Fetch and Store example 

* Fetch 
    1. Load the address into the MAR
    2. Decode the address in the MAR
    3. Copy the contents of that memory location into the MDR

* Store (address, value)
    1. Load the address into the MAR
    2. Load the value into the MDR
    3. Decode the address in the MAR
    4. Store the contents of the MDR into that memory location.

* The operation "decode the address in the mar" means that the memory unit must translate the N-bit binary address stored in the MAR into the set of signals need to access that one specific memory cell.


#### Cache Memory

* One bottleneck of computers today is that processors are able to execute instructions faster then then memory can be delivered to them from the RAM.

* When a program fetches a piece of data or an instruction, there is a likelyhood that:
    1. It will access that same instruction or piece of data in the very near future
    2. It will soon access the instructions or data that are located near that piece of data, where "near' means an address whose numerical value is close to this one.

* This observation is called the principle of locality.
    * When a computer uses something, it will probably use it again very soon, and will probably use the "neighbors" or this item very soon. (like a loop)

* This lead to the development of cache memory.

* A cache is typically 5 to 10 times faster than RAM but much smaller in size.

#### Input/Output and Mass Storage

* Input/output units are the devices that allow a computer system to communicate and interact with the outside world and store information. 

* RAM is a volatile memory which means all memory contained within it is lost when power is lost.

* Non volaatile memory will stay if power is lost.

* Input/Output devices are varied and range from a variety of things

* I/O access methods and I/O controllers.

* Input/output devices come in two basic types: those that represent information in human-readable form for human consumption those that store information in machine-readable form for access by a computer system.

* Direct access storage devices (DASDs) 

* Sequential access storage device (SASD)

* A direct access storage device may have different times when recieving data.

DASD examples:

* CDs, DVD, Hard drive.

* Seek time - time needed to position the read/write head

* Latency is the head and transfer time is the time for the entire sector to pass under the read/write head and have its contents read into or written from memory.



#### Arithmetic/Logic Unit

* The ALU is the subsystem that performs such mathematical and logical operations as addition, subtraction, and comparison form equality. 

* The ALU is made up of three parts: The registers, the interconnections between components, and the ALU circuitry. Together these components are called the data path.

* A register is a storage cell that holds the operands of an arithmetic operation and that, when the operation is complete, holds its result.

registers differ from RAM in several ways: 

* They do not have a numeric memory address but are accessed by a special register designator.

* The can be accessed much more quickly than regular memory cells. There are typically a few dozen to a houndred registers.

* They are not used for general-purpose storage but for specific purposes such as holding the operands for an upcoming arithmetic computation.

##### The Control Unit

* The most fundamental characteristic of the Von Neumann architecture is the stored program. A sequence of manchine language instructions stored as binary values in memory.

* It is the task of the control unit to: 

    1. Fetch from memory the next instruction to be executed.

    2. Decode it- determine what is to be done,

    3. Execute it by issuing the appropriate command to the ALU, memory, or I/O controllers. 

    * Machine Language Instructions are instructions that can be decoded and executed by the control unit of a computer.

    * Instructions in this language are expressed in binary.


1. Operation Code

2. Address Field 1

3. Address Field 2

4. ....

The operation code (op code) is a unique unsigned integer code assigned to each machine language operation.

The address field(s) are the memory addresses of the values on which this operation will work. 
    * If our computer has a maximum of 2^n memory cells, then each address field must be N bits wide to enable us to address every cell because it takes N binary difits to represent all adresses in the range 0 to 2^n

* The set of all operations that can be executed by a processor is called its instruction set.
