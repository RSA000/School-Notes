### Chapter 7: The internet, and the World Wide Web

* Computer network - computers connected together for the purpose of exchanging information.

* A computer network is a set of independent computer systems connected by telocommunication links for the purpose of sharing information adn resources.

* Individual computers on a network are referred to as nodes or hosts. 

#### Communication Links

* Communication links used to build a network vary widely in physical characteristics, error rate, and transmission speed. 

* Early on the most common way to transmit data was via switched, dial-up telephone lines.

* For binary signals to be transmitted via a switched dial-up telephone line, the signal must be restructured into the analog representation. A modem accomplishes this tasks. 

* A modem modulates or alters, a standard analog signal called a carrier wave so that it encodes binary information.

* The modem modifies the amplitude and frequency so that it is in one of two distinct states, 0 and 1.

* A modem can also demodulates a digital signal into an analog.

* A technology called broadband is rapidly replacing modems and analog phone lines. 

* Broadband generally refers to any communication link with a transmission rate exceeding 256,000 bps.

* There are currently two braodband options - digital subscriber line (DSL) and cable modem.

* A digital subscriber line uses the same wires that carry regular telephone signals into your home.

* A cable modem makes use of the links that deliver cable TV signals. Offers slightly higher speed then DSL.

* The most widely used broadband technology in commercial and office environments is ethernet.

* Ethernet was developed in the mid-1970s by computer scientists at the Xerox PARC research center.

* Wireless data communications uses radio, microwave, and infared signals. Wireless data communication was a breakthrough because it allowed a user to not be physically connected to a wire.. 

* There are two forms of wireless data communication.

* Wireless area netwwork (WLAN) A user transmits from his or her computer to a local base station, often referred to as a wireless router that is no more than a few hundred feet away. This base station is then connected to a traditional wired network.

* One of the omst widely used standards for wireless local access is called WIFI (for wireless fidelity)

* Wifi is used to connect a computer to the internet when it is within range (typicall 150-300 feet or 45-90 meters).

* Wifi systesm generally use the 2.4 GHz frequencies (5ghz is becoming the standard now.) and supports download transmission speeds of about 10-50mbs

* Bluetooth is a low-power wireless standard used to communicate between devices located close, usually within 30-50 feet (10-15 meters) 

* A Wireless wide area network (WWAN) big cellular antennas! (Think phone data, 4g, 5g...)

* Microwaves are line of sight, meaning the trabel in a straight line only. Because earth curves this is a problem. (range 15-80 kilometers)



##### Local Area Networks

There are two types of computer networks. A local area network (LAN) connects hardware devices such as computers, printers, and storage devices that are all in close proximity.

* An important characteristic of a LAN is the owners of the computers is also the owner of the means of communications. 

* Nodes of a LAN can have different topologies.
    * Nodes must "share the line" as intercepting information can cause problems
    * Some topologies are 
        * Bus - In a straight line
        * Ring - Like a circle
        * Star - Like a star - central node can route information

* There are two ways to construct an Ethernet LAN.
    * Shared cable - A wire is strung around through a building. Users can tap into the cable at the nearest point using a device call a transceiver.

    * Ether cables have a maximum allowable length.

    * A repeater is a device that simply amplifies and forwards a signal. 

    * A bridge is a "smarter" device that has knowledge about the nodes located on each separate network.

    * A switch contains a number of ports with a wire leading from each port to an Ethernet interface in the wall of a room in a building, or to a wireless router somewhere.






#### Wide Area Networks

* A wide area network (WAN) connects devices that are not in a close proximity but rather are across a large area.

* Because WANs cross public propert, users must purchase telecommunications services from an external provider.

* Typically there are dedicated point-to-point lines that directly connect two machines, not the shared channels like LAN.

* Most WANs use a store-and-forward, packet-switched technology. 

* One of the features of a store-and-forward network is that the failure of a single line or node does not bring down the entire network. 

#### Overal Structure of the Internet

* A router transmits messages between two distinct networks. Unlike a bridge, which connects two identical types of networks, routers can transmit information between networks that use totally different communication techniques, like an interpreter.

* An Internet service provider (ISP) is a business whose purpose is to provide access from a private network to the internet or from an individuals computer to the Internet. This access occurs through a WAN owned by the ISP. 

#### Communication Protocols 

* In networking, a protocol is a mutually agreed-upon set of rules, conventions, and agreements for the efficient and orderly exchange of information.

* The Internet is managed by the Internet Society, a nonprofit, nongovernmental, professional society composed of more than 100 worldwide organizations.

* The Internet Architecture Board (IAB) and the Internet Engineering Task Force (IETF), establishes and enforces network protocol standards.

* The Internet protocal hierachy, also called a protocol stack, has five layers. * This hierarchy is also referred to as TCP/IP, after the names of two of its most important protocols.


###### The five-layer TCP/IP protocol hierachy

5. Application: HTTP, SMTP, FTP

4. Transport: TCP, UDP

3. Network: IP

2. B Logical Link Control: PPP, Ethernet

2. A Medium access Control: Ethernet

1. Physical: Modem, DSL, Cable modem, Wi-Fi, 4G

###### Physical Layer

* The Physical layer protocols govern the exchqnge of binary digits across a physical communication channel, such as a fiber-optic cable, copper wire, or wireless radio channel. These protocols specify such things as:

* How we know when a bit is present on the line

* How much time the bit will remain on the line 

* Whether the bit is in the form of a digital or an analog signal

* What physical quantities are used to represent a binary 0 and a binary 1

* The shape of the connector between the computer and the transmission line.

* The goal of the Phusical layer is to create a "bit pipe" between two computers, such that bits put into the pipe at one end can be read and understood by the computer located at the other end. 

###### Data Link Layer 

* Because of various errors that can occur when transmitting information, error detection and correction is a problem that needs to be solved. 

* We also need to deal with framing, which is knowing what bits belong where.

* Error handling and framing is the process of two stages called layer 2a, Medium access control and layer 2b Logical link control. These two services form a layer 2 protocol called the Data Link Layer.

* The Medium Access Control protocols determine how to arbitrate ownership of a shared communication line when multiple nodes want to send messages at the same time.
    * Most Medium Access Control protocols including Ethernet, use a contention-based approach in which there is no central authority and all nodes compete equally for ownership of the line.
        * When a node wants to send a message, it first listens to the line to see whether or not it is currently in use. If the line is idle then the node transmits immediately. If the line is busy the node wanting to send monitors the status of the line and, as soon as it becomes idle, it transmits 

* A collision is when two messages are transmitted at exactly the same time.

* When a collision occurs, all information is lost. 

* When a collision occurs, the colliding nodes immediately stop sending, wait a random amound of time, and then attempt to resend.

* One reason the Ethernet protocol is so popular is that control is distributed. 

* It is the job of Layer 2b Logicial Link Control protocols to solve the error detection and correction problem and ensure that the message traveling across this channel from source to destination arrives correctly.

* We cant eliminate errors, but we can detect when an error has occurred and retransmit a new copy of the original message. 

* The ARQ algorithm, for automatic repeat request, is the basis for all Data Link Control protocols in current use.

* Start of packet - SOP, end of packet - EOP

SOP - sequence-number - message - error-check - EOP


###### Network Layer

* A great majority of nodes are not directly connected. It is the job of the end-to-end Network layer protocols to devliver a message from the site where it was created to its ultimate destination. To accomplish this delivery task, every node must agree to use the same addressing scheme so that everyone is able to idnetify that ultimate destination.

The two critical responsibilites of the network layer are:

* Creating a universal addressing scheme for all network nodes

* Delivering messages between any two nodes in the network.

* Every node in the network must run the identical Network layer protocol, and it is one of the most important parts of the protocol stack.

* The Network layer in the internet is called IP, for Internet Protocol

* Host names (example: website.com) are not the acutal names that nodes use to identify each other in IP. Nodes identify each other using a 32-bit IP address, often written as four 8-bit numeric quantities in the range of 0-255, each grouping separated by a dot. 

example: IP - 141.140.1.5 - 10001101 100001100 00000001 000000101

* It is the task of a special internet application called the Domain Name System (DNS) to convert from a symbolic host name to its equivalent 32-bit IP address. The DNS is a massive database, distributed over literally thousands of machines. 


* The process of selecting one specific path is called routing.

* The Internet uses a mehtod called Dijkstras shortest path algorithm.

###### Transport Layer

* Every computer has an IP address that uniquely identifies it.

* When a message comes in, how do we know which application program it is for and where to deliver it?

* We need a second level of address that identifies not only a specific machine but also a specific program running on that machine. This "program identifier", usually just a small integer value, is called a port number and it serves the same role as an address line.

* Assigning port numbers to programs and remembering which program goes with which port is part of the transport layer protocols.

* IP uses what is called a good faith transmission model. That means that it tries very hard to deliver a message form source to destination, but it does not guarantee delivery. 

* On the Internet, the primary transport protocol is TCO (Transport Control Protocol)

* TCP requires that the two programs at the source and destination node initially establish a connection. That is, the must first inform each other of the impeding message exchange, and they must describe the "quality of service" they want to receive. This connection does not exist in a hardware sense. It is a logical connection that exists only as entries in tables.

* TCP uses the same ARQ algorithm described previously.

###### Application Layer

* The application layer protocols are the rules for implementing the end user services provided by a network.

* HTTP - Hypertext Transfer Protocol - Web pages.

* SMTP - Simple Mail Transfer Protocol - Sending electronic mail

* A single web page is identified by a symbolic string called a Uniform Resource Locator (URL).

* URLS have three parts:

protocol://host address/page

* The first part, protocol indicates the type of information contained in this page. The most common format is hypertext and we can access it using the Hypertext Transfer Protocol (HTTP).

* The second part of the URL is the host address of the machine where the page is stored. This is the symbolic host name discussed earlier

* The third part of the URL is the page identification which is usually a file stored on the specified machine.
