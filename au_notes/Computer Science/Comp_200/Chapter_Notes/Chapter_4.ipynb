#### The Binary Numbering System.

* There are two types of information representation: External and internal

* External would be how to a computer would display information on a screen and internal is how the computer would store and use information in itself.

* Computers use the binary numbering system which consists of 0's and 1's (binary = two) and the positions of these values.

* Binary is a base-2 positional numbering system.

* From the right to left the binary system of counting works like: 2^0 is ones 2^2 is twos and so on.

* To internalize decimal numbers in a machine readable form, we use the decimal-to-binary algorithm, which is based on successive division by 2. Dividing the original decimal value by 2 produces a quotient and a remainder, which must be either a 0 or a 1. Record the remainder digit then divide the quetient by 2, getting a new quotient and a second remainer digit. The process of dividing by 2, saving the quotient, and writing down the ramainder is repeaeted until the quotient equals 0. The sequence of remainder digits, when written left to right from the last remainder digit to the first, is the binary representation of the original decimal value.


* Every computer has a maximum number of binary digits that can be used to store an integer. 
    * These values are usually 16, 32, or 64 bits.

* Depending on the maximum size, we can determine the maximum integer a computer can store. (16 = 1111111111111111 = 65,535) 

* Computer science also uses octal (base 8) and hexadecimal (base 16)

* Any operation on a computer that produces an unsigned integer larger than its maximum representation results in an error called arithmetic overflow.

* Arithmetic overflow is an attempt to represent an integer that excees the maximum allowable value.

##### Signed Numbers. 

* Binary digits can also represent signed integer (positive and negative).

* A common form of representing signed integers is having the left most digit represent whether an integer is negative or positive.

* A common form of this is the twos complement representation where zero is 000... and -1 is 100.. which is used to cut out a "negative zero". This way the negative and postive numbers are divided "evenly" (the negative would have one extra ).

How does a computer know if the integer is signed or unsigned? The answer is that a computer does not know. A sequence of binary digits can have many different interpretations, and there is no fixed, predetermined interpretation given to any binary value. A Binary number stored in memory of a computer takes on a meaning only because it is used in a certain way.

#### Binary Representation of Sound and Images

* Sound is analog information.

In an analog representation, objects can take on any value. In the case of sound, a tone is a continuous sinusoidal waveform that varies in a regular periodic fashion over time.

* The amplitude (height) of the wave is the measure of its loudness. The greater the amplitude, the louder the sound. 

* The period of the wave, designated as T, is the time it takes for the wave to make one complete cycle. 

* The frequency f is the total number of cycles per unit time measured in cycles/second, also called hertz.

* f = 1/T

* The frequency is a measure of the pitch, the highness or lowness of a sound. The higher the freqency the higher the perceived tone.

* Humans can generally detect sounds in the range of 20 to 20,000 hertz.

To store a waveform in a computer, it must first be digitized  (converted to a digital representation).

* Sampling is measuring and storing soundwaves as information at fixed intervals. The amplitude of the signal is measured and stored as a sequence of sampled numerical amplitudes.

* Increased sampling (taking more samples in a smaller amount of time) increases the authenticity of the original analog sound. This is known as sampling rate.

* A sampling rate of R samples/second allows to reproduce all frequencies up to about R/2 hertz. Because the human ear can normally detect sound of to about 20,000 hertz, a sampling rate of at least 40,000 samples per second is necessary to capture all audible frequencies.

* Bit depth is the number of bits used to encode each sample.

* Common audio-encoding formats: WAV (Waveform Audio File Format), AAC (Advanced Audio Coding), WMA (Windows Media Audio), and MIDI (Musical Instrument Digital Interface). MP3 (MPEG-1 Audio Level 3 Encoding)

The base two system is much more reliable this way.

 #### Boolean Logic and Gates

The construction of computer circuits is based on the branch of mathematics and symbolic logic called Boolean logic.

* This area of mathematics deals with rules for manipulating the two logical values true and false.

* It is used to construct circuits that perform operations such as adding numbers, comparing numbers, and fetching instructions. 

* These ideas are part of the branch of computer science known as hardware design, also called logic design

* Boolean expression: Any expression that evaluates to either tru or false.

* Boolean expression are AND, OR, and NOT.

* We use Boolean expression to create logic gates and we use logic gates to construct circuits.

* A circuit is a collection of logic gates that transforms a set of binary inputs into a set of binary outputs.


#### Control Circuits

Control circuits are used to determine the order in which operations are carried out and to select the correct data values to be processed.

* The two major types of control circuits are called multiplexors and decoders.

* A multiplexor is a circuit that has 2n inputs and 1 output line. Its function is to select exactly one of its 2n input lines and copy the binary value on that input line onto its single output line.

* A multiplexor chooses one specific input by using an additional set of n lines called selector lines.
    * multiplexors have 2^n input lines, n selector lines and one output.

* A decoder works the opposite of the multiplexor.
    * decoders have n input lines and 2^n output. 

* Together, decoder and multiplexor circuits enable us to build computer systems that execute the correct instructions using the correct data values.
