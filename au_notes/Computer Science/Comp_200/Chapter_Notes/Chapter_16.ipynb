#### Computer Graphics and Entertainment: Movies, Games, and Virtual Communities


#### Computer-Generated Imagery (CGI)


##### The Graphics Pipeline

* Object modeling

* Object motion

* Rendering and display

#### Object Modeling

* The first step in generating a realistic three-dimensional image is object modeling-the creation of a mathematical or computational model of a three-dimensional object that can be stored in the computer's memory and maipulated algorithmically.

* There are moany appraoches to object modeling, but one of the most well known and widely used is wireframe modeling.
    * In this technique, the object's surface, but not its interior, is represented mathematically using a set of simple nonoverlapping polygons, usually triangles or rectangles.

* Using special CGI software and an algorithm called tessellation, the image is subdivided into a set of plane figures that completely covers its surface.


#### Object Motion

* After we have captured and stored a model of our object in a computational format, we can begin the next stage in the graphics pipeline.

* There are three types of rigid motion (motion that does not bend or deform an object): translation, rotation, and reflection.

    * tranlsation - moving around 

    * rotation - spinning

    * reflection - mirroring

* To implement these three motions, we use a mathematical structure called a transformation matrix. When a vector contining the (x, y, z) coordinates of a single vertex point is multiplied by this matrix, the result is a new vector containing the translated, rotated, or reflected (x\`, y\`, z`)


#### Rendering 

* Displaying the final image.

* Rednering means taking an object stored as a mathematical model, such as the vertex list, and converting it into a fully formed, visually pleasing three-dimensional image.

* Rendering is a complex set of operations that often consumes the great majority of computer time required to produce an image. Some of the issues addressed during the rendering process are:

* Lighting
* Color shading
* Shadows
* Texture mapping
* Blur

#### Video Gaming

* A movie is not an interactive environment. There is no change to the plot or action on the screen.

* A movie is a static environment that is created once and shown as often as desired.

* A video game is a highly interactive environment. Using an input device such as a joystick, wireless controller, or finger tap, a user dynamically controls the action and makes instant decisions about what happens next.

* The branch of computer graphics that studies methods for creating images at a rate matching that of the real world is called real-time graphicss.

* If necessary, sacrifice image quality for speed of display.

* One of the most common techniques for increasing imaging speed, term the frame rate, is to use a GPU, an acronym for Graphics Processing Unit.

* A GPU is an independent Von Neumann processor. 

* A GPU executes instructions in parallel with the CPU, the Central Processing Unit, and carries out all the graphics operations described.

* Typically, a GPU has its own dedicated RAM. 

* The GPU along with dedicated RAM, referred to as video memory, is located on a video card, also called a graphics card, and is connected to the main CPU and memory through plug-in expansion slot or via the system bus.

* Another technique to speed up rendering and display is culling. Rather then rendering every plane in the wireframe model and then determining which planes are visible from the user's perspective, we could turn those two operations around. First determine which planes can be seen from the user's point of view, based on location and opaqueness, and then render only those objects visible in the next frame, omitting all operations on hidden surfaces. 

* Video gaming often makes use of a technique called cut-ins. These are fully modeled and fully rendered objects stored in a video library in video memory. These already prepared objects can be dropped into a frame as is, producing a significant speedup in frame creation. These cut-ins often include images of the main game characters as well as standard back-ground objects.
