##### Misconceptions About Computer Science
---
1. Computer Science is the study of computers:
    * This is considered an incomplete statement. Computer science has existed before actual computers.
    * Some of the earliest fundamental theoretical work on computer science was done in the 1920's-1940's.
    * Before, computer science was considered to be a branch of logic and applied mathematics.
    * Computer science wasn't considered its own field until the late 1950's/early 1960's.
    * "Computer Science is no more about computers than astronomy is about telescopes, Biology about microscopes or Chemistry is about beakers and test tubes. Science is not about tools. It is about how use them and what we find out when we do." - Micheal R. Fellows and Ian Purberry.
---
2. Computer Science is the study of how to write computer programs.
    * Typically, people begin to learn about computer science by learning a programming langauge which can create misunderstanding about what computer science is. 
    * While programming is an essential skill for a computer scientists (for testing and implementing ideas) Computer science is more about the methods used to build programs rather then the actual programs themselves.
---
 3. Computer Science is the study of the uses and applications of computers and software.
    * Learning to use software is no more a part of computer science than drivers education is a branch of automotive engineering.
    * A computer scientists job is specifying, designing, building and testing software paclages as well as the computers on which they run.

##### Definition Of Computer Science

Professors Norman Gibbs and Allen Tuckers definition of a computer scientist.

* The task of a computer scientist is to design and develop algorithms.

The Design process is as follows: 

1. Studying the behavior of algorithms to determine if they are correct and efficient. - Their formal and mathematical properties.

2. Designing and building computer systems that are able to execute algorithm. - Hardware realizations

3. Designing programming langauges and translating algorithms into these langauges so they can be executed by the hardwares. - Linguistic realizations

4. Identifying important problems and designing correct and efficient software packages  to solve these problems - Applications

##### Definition Of Algorithm

* A procedure for solving a mathematical problem in a finite number of steps that frequently involves repetition of an operation: Broadly; a step-by-step method for accomplishing some task.

* A well-ordered collection of unambiguous operations that, when executed, produces a result and halts in a finite amount of time.

* When an operation is unambiguous without further simplication or explanation, we call it a primitive operation.

##### Common Operations

1. Sequential - Carries out a single well-defined task.

2. Conditional - "Question asking" instructions of an algorithm.

3. Iterative - "Looping instructions", Repeating execution of previous block of instructions

##### A Brief History Of Computing 

###### Early Period: Upto 1940's 

* 3,000 years ago, Greeks, Babylonians, Indian, Chinese, and Persians where interested in and contributed to the field of mathematics.

* SCotsman John Napier invented logarithms as a way to simplify difficult mathematical computations.

* The first slide rule appeared around 1622

* 1672 - French philosopher and Mathematician Braise Pascal designed and built one of the first mathematical calculators (Named the Pascaline). The calculator was completely mechanical and could add and subtract.

* 1674 - A German mathematician Gottfried Leibnitz created another mechanical calculator that could add, subtract, multiply and divide. Named the Liebnitz Wheel.

* The two machines described aboved used interlocking wheels and gears to store numbers and perform basic arithmatic. 

* Although the Pascaline and Liebnitz wheel where wonders for their time, they lacked certain features that would make them computers:

    1. They did not have a memory where information could be stored in machine readable form. 

    2. The where not programmable. A person could not provide in advance a sequence of instructions that could be executed by the device without manual intervention.

##### The First "Computer Device"

The First actual computing device was an automatic loom. In 1801, Frenchman Joseph Jacquard invented a loom that satisfied all the criteria of what a computer is. Joseph wanted to automate the weaving process which at the time was very time consuming. The loom used punched cards to create a desired pattern.

#### 1946 - John Von Neuman

Von Neuman was an integral component to modern day computers.

* Von Neuman was a genius and one of the greatest mathematicians of our time.

* Von Neuman was a mathematician, experimental physicist, chemist, economist, and computer scientist.

* Taught at Princeton University of Pennsylvania and worked with Mauchly and Eckert on the ENIAC.

* Developed a model known as the stored program computer.
    * Until this model, all computers were programmed externally using wires, connectors and plugboards. The memory of these computers only stored data and not instructions.

* Von Neumann proposed that the instructions that control the operation of the computer could be encoded as binary values and stored internally in the machine. 

* Von Nuemann invented programming as it is known today.

* The model of computing proposed by Von Neumann included many orhter important features found on all modern computing systems.

* This model of computation is called Von Neumann architucture


####  EDVAC, EDSAC

#### The Modern Era: 1950's to the present.

* At this point, computer systems development has primarily been an evolutionary process instead of a revolutionary one.

* The overall structure of a computer did not change much, but breakthroughs where being made in memory, speed, size, and cost.

##### First Generation: 1950-1957

* Is considered the first generation of computing.
* The UNIVAC 1 was the first commercial computer and was built for sale.
* The IBM 701 was built.
* Computers at this time used vacuum tubes for processing and storage and where extremely difficult to maintain and also consumed a lot of power.
* The first generation of computers where only operated by trained personnel and only in specialized applications such as large corporations, military, universities, and research labs.


###### Second Generation: 1957-1965 

The second generations of computing brought in a major change in the size and complexity of computers. Computers moved from vacuum tubes to transistors which greatly reduced their size and power consumption. Memory was now being made using tiney magnetic cores. These innovations made computers less costly and allowed for smaller organizations to buy and use computers.

The first high level programming langauge was created in 1957 which curved the difficulty of programming a computer. This created the occupation of programming. 

###### Third Generation: 1965-1975

* The third generation of computers introduced integrated circuits rather then using discrete electronic components.

* Integrated circuits with transistors, resistors, and capacitors were photographically etched onto a piece of silicon, reducing the cast and size of computers. Computers went from building sized, to room sized, to desk sized.

* This period of time was also the birth of the minicomputer.
    * The PDP-1 manufactured by the Digital Equipment Corp.

* The birth of the software industry. 
    * Companies specializing in designing and building programs for these computers sprang up to assist in making computers more accessible by regular people.

* By the 1970's, computers where much more common place.

##### 1975-1985: Fourth Generation

* First appearence of the microcomputer.

* Intergrated circuit technology advances to the point where a single circuit board can contain and entire computer system. Computers are now the size of a typewriter.

* The worlds first microcomputer was the Altair 8800. 

* Programs such as spreadsheets, databses, word processors, and drawing programs.

* Birth of the first computer networks.

* Electronic mail.

* User friendly programs.

* Embedded systems - devices that contain a computer to control their internal operation.

2. Heuristics
---
A heuristic technique is an approach to solving a problem that doesn't necessarily have to be optimal or perfect.

This technique differs from an algorithmic approach as algorithms are supposed to be efficient and accurate while the heuristic technique is more getting a "ballpark range".

For example and algorithm for addition of a three digit numbermay be:
    1. Start with the 1's position
    2. If the addition of the two 1's place numbers is greater than 10, add 1 to the 10's place and place the remainder in the 1's result
    3. Move to the 10's place
    4. If the addition of the two 10's place numbers is greater than 10, add 1 to the 10's place and place the remainder in the 10's result.
    5. Move to the 100's repeat step 4 except this time it's 100's instead of tens.

A Heuristic method may go like this:

    1. Start at the 10's place
    2. If the addition of the two 10's place numbers is greater than 10, add 1 to the 10's place and place the remainder in the 10's result.
    3. Move to the 100's repeat step 2 except this time it's 100's instead of tens.

This method wouldn't be completely accurate but would still give a good approximation of what the answer would be and takes less steps.
